```python
import pandas as pd
import numpy as np
```

# 1.选取Series数据

```python
#读取college数据集，查看CITY前5行
college = pd.read_csv('data/college.csv',index_col='INSTNM')
city = college['CITY']
city.head()
```

```python
#iloc可以通过整数选取
city.iloc[3]
```

```python
#iloc通过整数列表选取多行，返回结果是Series
city.iloc[[10,20,30]]
```

```python
#选择等分的数据，可以使用切片语法
city.iloc[4:50:10]
```

```python
#loc只接收行索引标签
city.loc['Heritage Christian University']
```

```python
#随机选择4个标签
np.random.seed(1)
labels = list(np.random.choice(city.index,4))
labels
```

```python
#通过标签列表选择多行
city.loc[labels]
```

```python
#也可以通过切片语法均匀选择多个
city.loc['Alabama State University':'Reid State Technical College':10]
```

```python
#也可以不使用loc，直接使用类似Python的语法
city['Alabama State University':'Reid State Technical College':10]
```

```python
#要想只选取一项，并保留其Series类型，则传入一个只包含一项的列表
city.iloc[[3]]
```

```python
#使用loc切片时要注意，如果start索引再stop索引之后，则会返回空，并且不会报警
city.loc['Reid State Technical College':'Alabama State University':10]
```

```python
#也可以切片逆序选取
city.loc['Reid State Technical College':'Alabama State University':-10]
```

# 4.用整数和标签选取数据

```python
#读取college数据集，行索引命名为INSTNM
college = pd.read_csv('data/college.csv',index_col='INSTNM')
#用索引方法get_loc，找到指定列的整数位置
col_start = college.columns.get_loc('UGDS_WHITE')
col_end = college.columns.get_loc('UGDS_UNKN') + 1
col_start,col_end
```

```python
#用切片选取连续的列
college.iloc[:5,col_start:col_end]
```

```python
#index()方法可以获得整数行对应的标签名
row_start = college.index[10]
row_end = college.index[15]
college.loc[row_start:row_end,'UGDS_WHITE':'UGDS_UNKN']
```

# 5.快速选取标量

```python
#通过将行标签赋值给一个变量，用loc选取
college = pd.read_csv('data/college.csv',index_col='INSTNM')
cn = 'Texas A & M University-College Station'
college.loc[cn,'UGDS_WHITE']
```

```python
#at可以实现同样的功能
college.at[cn,'UGDS_WHITE']
```

```python
#用魔术方法%timeit，对速度进行比较
#.iat和.at只接收标量值，是专门用来取代.iloc和.loc选取标量的，可以节省大概2.5微秒
%timeit college.loc[cn,'UGDS_WHITE']
```

```python
%timeit college.at[cn,'UGDS_WHITE']
```

```python
#用get_loc找到整数位置，再进行速度比较
row_num = college.index.get_loc(cn)
col_num = college.columns.get_loc('UGDS_WHITE')
row_num,col_num
```

```python
%timeit college.iloc[row_num,col_num]
```

```python
%timeit college.iat[row_num,col_num]
```

```python
%timeit college.iloc[5,col_num]
```

```python
%timeit college.iat[5,col_num]
```

```python
#Series对象也可以使用.iat和.at选取标量
state = college['STABBR']
state.iat[1000]
```

```python
state.at['Stanford University']
```

# 6.惰性行切片

```python
#读取college数据集；从行索引10到20，每隔一个取一行
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college[10:20:2]
```

```python
#Series也可以进行同样的切片
city = college['CITY']
city[10:20:2]
```

```python
#查看第4002个行索引标签
college.index[4001]
```

```python
#Series和DataFrame都可以用标签进行切片。下面是对DataFrame用标签切片
start = 'Mesa Community College'
stop = 'Spokane Community College'
college[start:stop:1500]
```

```python
#下面是对Series用标签切片
city[start:stop:1500]
```

```python
#惰性切片不能用于列，只能用于DataFrame的行和Series，也不能同时选取行和列
#下面尝试选取两列，导致错误
college[:10,['CITY','STABBR']]
```

```python
#只能用.loc和.iloc选取
first_ten_instnm = college.index[:10]
college.loc[first_ten_instnm,['CITY','STABBR']]
```

# 7.按照字母切片

```python
#读取college数据集；尝试选取字母顺序在'Sp'和'Su'之间的学校
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college.loc['Sp':'Su']
```

```python
#对college进行排序
college = college.sort_index()
#college = college.head()
college
```

```python
#再次尝试选取字母顺序在'Sp'和'Su'之间的学校
pd.options.display.max_rows = 6
college.loc['Sp':'Su']
```

```python
#可以用is_monotonic_increasing或is_monotonic_decreasing检测字母排序的顺序
college = college.sort_index(ascending=False)
college.index.is_monotonic_decreasing
```

```python
#字母逆序选取
college.loc['E':'B']
```
