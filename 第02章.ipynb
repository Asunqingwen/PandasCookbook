# 6.比较缺失值

```python
#Pandas使用NumPy NaN(np.nan)对象表示缺失值，这是一个不等于自身的特殊对象。
np.nan == np.nan
```

```python
#Python的None对象是等于自身的
None == None
```

```python
#所有和np.nan的比较都返回False，除了不等于
5 > np.nan
```

```python
np.nan > 5
```

```python
5 != np.nan
```

```python
# college_ugds_所有值和.0019比较，返回布尔值DataFrame
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head() == .0019
```

```python
#用DataFrame和DataFrame进行比较
college_self_compare = college_ugds_ == college_ugds_
college_self_compare.head()
```

```python
#用all()检查是否所有的值都是True；这是因为缺失值不互相等于
college_self_compare.all()
```

```python
#可以用==号判断，然后求和
(college_ugds_ == np.nan).sum()
```

```python
#统计缺失值最主要的方法是使用isnull方法
college_ugds_.isnull().sum()
```

```python
#比较两个DataFrame最直接的方法是使用equals()方法
from pandas.testing import assert_frame_equal
try:
    print(assert_frame_equal(college_ugds_,college_ugds_))
except:
    print('test')
```

```python
#eq()方法类似于==，和前面的equals有所不同
college_ugds_.eq(.0019).head()
```

# 7.矩阵转置

```python
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
#count()返回非缺失值的个数
college_ugds_.count()
```

```python
#axis默认设为0
college_ugds_.count(axis=0)
```

```python
#等价于axis='index'
college_ugds_.count(axis='index')
```

```python
#统计每行的非缺失值个数
college_ugds_.count(axis='columns').head()
```

```python
#除了统计每行的非缺失值个数，也可以求和加以确认
college_ugds_.sum(axis='columns').head()
```

```python
#用中位数了解每列的分布
college_ugds_.median(axis='index')
```

```python
#使用累积求和cumsum()可以很容易看到白人、黑人、西班牙裔的比例
college_ugds_cumsum = college_ugds_.cumsum(axis=1)
college_ugds_cumsum.head()
```

```python
#UGDS_HISP——列降序排列
college_ugds_cumsum.sort_values('UGDS_HISP',ascending=False,axis=0).head()
```

# 8.确定大学校园多样性

```python
# US News给出的美国10所最具多样性的大学
pd.read_csv('data/college_diversity.csv',index_col='School')
```

```python
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
college_ugds_.isnull().sum(axis=1).sort_values(ascending=False).head()
```

```python
#如果所有列都是缺失值，则将其去除
college_ugds_ = college_ugds_.dropna(how='all')
college_ugds_.isnull().sum()
```

```python
#用大于或等于方法ge()，将DataFrame变为布尔值矩阵
college_ugds_.ge(.15).head()
```

```python
#对所有True值求和
diversity_metric = college_ugds_.ge(.15).sum(axis='columns')
diversity_metric.head()
```

```python
#使用value_counts()，查看分布情况
diversity_metric.value_counts()
```

```python
#查看哪些学校种群比例超过15%的数量多
diversity_metric.sort_values(ascending=False).head()
```

```python
college_ugds_.loc[['Regency Beauty Institute-Austin','Central Texas Beauty College-Temple']]
```

```python
#查看US News前五所最具多样性的大学在diversity_metric中的情况
us_new_top = ['Rutgers University-Newark',
             'Andrews University',
             'Stanford University',
             'University of Houston',
             'University of Nevada-Las Vegas']
diversity_metric.loc[us_new_top]
```

```python
#可以用最大种群比例查看哪些学校最不具有多样性
college_ugds_.max(axis=1).sort_values(ascending=False).head(10)
```
